Also, price and odometer columns are numeric data stored as text. These will need to be changed.

```python
autos["price"] = autos["price"].str.replace("$","")
autos["price"] = autos["price"].str.replace(",","")
autos["price"] = autos["price"].astype(int)
autos["price"] = autos["price"].round(3)
autos["odometer"] = autos["odometer"].str.replace("km","")
autos["odometer"] = autos["odometer"].str.replace(",","")
autos["odometer"] = autos["odometer"].astype(int)
autos["odometer"] = autos["odometer"].round(5)
autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
```

```python
top_car_count = autos["brand"].value_counts().head(20).index
```

I have chosen to aggregate on the top 20 most popular brands. The working is below.

```python
top_20_cars_price = {}
for car in top_car_count:
    selected_rows = autos[autos["brand"] == car]
    mean_price = selected_rows["price"].mean()
    top_20_cars_price[car] = mean_price
```

```python
print(top_20_cars_price)
```

```python
top_20_cars_miles = {}
for car in top_car_count:
    selected_rows = autos[autos["brand"] == car]
    mean_mileage = selected_rows["odometer_km"].mean()
    top_20_cars_miles[car] = mean_mileage
```

```python
print(top_20_cars_miles)
```

```python
bmp_series = pd.Series(top_20_cars_price)
print(bmp_series)
```

```python
top_mileage_series = pd.Series(top_20_cars_miles)
print(top_mileage_series)
```

```python
top_20_cars = pd.DataFrame(top_mileage_series, columns=['mean_mileage'])
top_20_cars["mean_price"] = pd.DataFrame(bmp_series, columns=['mean_price'])

top_20_cars.sort_values("mean_price", ascending = False)
```

There seems to be a slight inverse proportional relationship between mileage and price. Sonstige autos have the highest mean price and the lowest mean mileage, while it is vice versa for volvo.
